```python
import pandas as pd
pd.options.mode.chained_assignment = None
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import random
import string
```

```python
y_df = pd.read_csv('../../data/model_results.csv')
```

```python
del y_df['success']
```

```python
y_df.group.unique()
```

```python
y_df['status'] = ''

y_df['status'][y_df['group']=='weekly-cancel'] = 'One Week Trial - Immediate Cancel'
y_df['status'][y_df.index < 10] = 'One Week Trial - Recent Cancel'

y_df['status'][y_df['group']=='monthly-cancel'] = 'One Month Trial - Immediate Cancel'
y_df['status'][(y_df['status']=='One Month Trial - Immediate Cancel') & 
               (y_df.index < 31000)] = 'One Month Trial - Recent Cancel'

y_df['status'][y_df['group']=='weekly-eligible'] = 'One Week Trial - Low P1 Probability'
y_df['status'][(y_df['group']=='weekly-eligible') & 
               (y_df.probs >= .25)] = 'One Week Trial - Medium P1 Probability'
y_df['status'][(y_df['group']=='weekly-eligible') & 
               (y_df.probs >= .60)] = 'One Week Trial - High P1 Probability'

y_df['status'][y_df['group']=='monthly-eligible'] = 'One Month Trial - Low P1 Probability'
y_df['status'][(y_df['group']=='monthly-eligible') & 
               (y_df.probs >= .25)] = 'One Month Trial - Medium P1 Probability'
y_df['status'][(y_df['group']=='monthly-eligible') & 
               (y_df.probs >= .60)] = 'One Month Trial - High P1 Probability'
```

```python
y_df.status.unique()
```

```python
import numpy as np
y_df['subscriber_id'] = np.random.randint(1, len(y_df), y_df.shape[0])
```

```python
y_df
```

```python
def random_char(y):
       return ''.join(random.choice(string.ascii_letters) for x in range(y))
    
y_df['email'] = [random_char(7) + '@gmail.com' for x in y_df.index]
```

```python
y_df.status.unique()
```

```python
y_df['intervention_cluster'] = ''
y_df['intervention_cluster'][y_df['status']=='One Week Trial - Immediate Cancel'] = 'No Usage'
update_df = y_df[y_df['status']=='One Week Trial - Immediate Cancel'].sample(frac = 0.2)
update_df['intervention_cluster'] = 'Converter Look-a-Like'
y_df.update(update_df)
update_df = y_df[y_df['status']=='One Week Trial - Immediate Cancel'].sample(frac = 0.6)
update_df['intervention_cluster'] = 'Low Usage: Difficult Prediction'
y_df.update(update_df)

y_df['intervention_cluster'][y_df['status']=='One Month Trial - Immediate Cancel'] = 'No Usage'
update_df = y_df[y_df['status']=='One Month Trial - Immediate Cancel'].sample(frac = 0.3)
update_df['intervention_cluster'] = 'Converter Look-a-Like'
y_df.update(update_df)
update_df = y_df[y_df['status']=='One Month Trial - Immediate Cancel'].sample(frac = 0.5)
update_df['intervention_cluster'] = 'Low Usage: Difficult Prediction'
y_df.update(update_df)

y_df['intervention_cluster'][y_df['status']=='One Week Trial - Recent Cancel'] = 'Medium Probability P1'
y_df['intervention_cluster'][y_df['status']=='One Month Trial - Recent Cancel'] = 'Medium Probability P1'

y_df['intervention_cluster'][y_df['status']=='One Week Trial - High P1 Probability'] = 'High Probability P1: 60 Min Usage'
y_df['intervention_cluster'][y_df['status']=='One Month Trial - High P1 Probability'] = 'High Probability P1: 60 Min Usage'

y_df['intervention_cluster'][y_df['status']=='One Week Trial - Low P1 Probability'] = 'Low Probability P1: Dormant'
y_df['intervention_cluster'][y_df['status']=='One Month Trial - Low P1 Probability'] = 'Low Probability P1: Dormant'

y_df['intervention_cluster'][y_df['status']=='One Week Trial - Medium P1 Probability'] = 'Medium Probability P1'
y_df['intervention_cluster'][y_df['status']=='One Month Trial - Medium P1 Probability'] = 'Medium Probability P1'
```
