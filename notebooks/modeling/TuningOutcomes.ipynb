```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv("../../saved_models/gridsearch_results_RandomForestClassifier_month.csv", index_col=0)
results
```

```python
results.columns
```

```python
max_depth = results.groupby(["param_max_depth"])\
    .mean()["mean_test_score"]\
    .reset_index()
max_depth
```

```python
min_samples_leaf = results.groupby(["param_min_samples_leaf"])\
    .mean()["mean_test_score"]\
    .reset_index()
min_samples_leaf
```

```python
min_samples_split = results.groupby(["param_min_samples_split"])\
    .mean()["mean_test_score"]\
    .reset_index()
min_samples_split
```

```python
n_estimators = results.groupby(["param_n_estimators"])\
    .mean()["mean_test_score"]\
    .reset_index()
n_estimators
```

```python
plt.plot(n_estimators["param_n_estimators"], n_estimators["mean_test_score"])
plt.xlabel("Number of decision trees")
plt.ylabel("Mean Precision")
plt.show()
```

```python
plt.plot(min_samples_split["param_min_samples_split"], min_samples_split["mean_test_score"])
plt.xlabel("Min samples to split")
plt.ylabel("Mean Precision")
plt.show()
```

```python
plt.plot(min_samples_leaf["param_min_samples_leaf"], min_samples_leaf["mean_test_score"])
plt.xlabel("Min samples for leaf")
plt.ylabel("Mean Precision")
plt.show()
```

```python
plt.plot(max_depth["param_max_depth"], max_depth["mean_test_score"])
plt.xlabel("Max Depth")
plt.ylabel("Mean Precision")
plt.show()
```
